```python
df=df.replace(to_replace ="First Owner",  value ="1") 
df=df.replace(to_replace ="Third Owner",  value ="3") 
df=df.replace(to_replace ="Second Owner",  value ="2") 
df=df.replace(to_replace ="Fourth & Above Owner",  value ="3") 
df=df.replace(to_replace ="Test Drive Car",  value ="3")
df=df.replace(to_replace ="Individual",  value ="1")
df=df.replace(to_replace ="Dealer",  value ="2")
df=df.replace(to_replace ="Trustmark Dealer",  value ="3")
df=df.replace(to_replace ="Manual",  value ="1")
df=df.replace(to_replace ="Automatic",  value ="2")
df=df.replace(to_replace ="Petrol",  value ="1")
df=df.replace(to_replace ="Diesel",  value ="2")
df=df.replace(to_replace ="CNG",  value ="3")
df=df.replace(to_replace ="LPG",  value ="4")
df=df.replace(to_replace ="Electric",  value ="5")
df.head(10)
```

```python
#feature importance
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(x,y)
```

```python
#highly important features
feat_importances=pd.Series(model.feature_importances_,index=x.columns)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
```
